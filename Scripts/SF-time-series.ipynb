```python
aqi_county_data_18['year'] = 2018 
aqi_county_data_16['year'] = 2016
aqi_county_data_14['year'] = 2014 


aqi_data = aqi_county_data_18.append(aqi_county_data_16)
aqi_data = aqi_data.append(aqi_county_data_14)
aqi_data.reset_index(level=0, inplace=True)
```

```python
# Standardizing Column names
# AQI County
aqi_county_data.columns = aqi_county_data.columns.str.replace(' ', '_') # replacing spaces (' ') with  underscore.
aqi_county_data.columns = aqi_county_data.columns.str.replace('-', '_') # replacing hyphens ('-') with underscore
aqi_county_data.columns= aqi_county_data.columns.str.lower() # converting column names to lower case
```

```python
# Standardizing Column names
# AQI
aqi_data.columns = aqi_data.columns.str.replace(' ', '_') # replacing spaces (' ') with  underscore.
aqi_data.columns = aqi_data.columns.str.replace('-', '_') # replacing hyphens ('-') with underscore
aqi_data.columns= aqi_data.columns.str.lower() # converting column names to lower case

aqi_data['state_name'] = aqi_data['state_name'].str.lower()
aqi_data['county_name'] = aqi_data['county_name'].str.lower()
```

```python
# Standardizing Column names
# svi_data
svi_data.columns = svi_data.columns.str.replace(' ', '_') # replacing spaces (' ') with  underscore.
svi_data.columns = svi_data.columns.str.replace('-', '_') # replacing hyphens ('-') with underscore
svi_data.columns= svi_data.columns.str.lower() # converting column names to lower case

svi_data['state'] = svi_data['state'].str.lower()
svi_data['county'] = svi_data['county'].str.lower()
```

```python
svi_data.columns
```

# Pre-process

```python
aqi_data_agg = aqi_data.groupby(['county_name', 'state_name', 'year'])['aqi'].agg('mean')
```

```python
aqi_data_agg= pd.DataFrame(aqi_data_agg)
aqi_data_agg.reset_index(level=0, inplace=True)
```

```python
aqi_data_agg.head()
```

```python
aqi_svi_master_data = pd.merge(aqi_data_agg,svi_data,how='left',left_on=['county_name','state_name','year'], right_on= ['county','state','year'])
```

```python
aqi_svi_master_data.head(50)
```

```python
aqi_svi_master_data.to_gbq(destination_table= 'aqi_data.aqi_svi_master_data', 
          project_id='cdss-datathon-327823', 
          if_exists='replace', 
          progress_bar=True)
```
